```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
# first, the string of list has to be converted into integers
# therefore i have used ast.literal_eval for changing the object into correct grammer
```

```python
#cast
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
                
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#forming vectors, using bag of words, using stemming
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:20]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list =sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:20]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
import tkinter as tk

# Define a function that takes an input and generates some output
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list =sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:40]
    recommended_movies = []
    for i in movies_list:
        recommended_movies.append(new_df.iloc[i[0]].title)
    return recommended_movies

# Create the main window
root = tk.Tk()

# Set the title of the window
root.title("Movie Recommender")

# Set the background color of the window
root.configure(bg='#2d2d2d')

# Create a label to prompt the user to enter a movie name
input_label = tk.Label(root, text="Enter a movie name:", font=("Arial", 14), fg='#ffffff', bg='#2d2d2d')
input_label.pack()

# Create an entry widget for the user to enter a movie name
input_entry = tk.Entry(root, font=("Arial", 14), width=30)
input_entry.pack(pady=10)

# Create a label to display the recommended movies
output_label = tk.Label(root, text="Recommended Movies:", font=("Arial", 14), fg='#ffffff', bg='#2d2d2d')
output_label.pack()

# Create a text box to display the recommended movies
output_text = tk.Text(root, font=("Arial", 14), height=10, width=50, fg='#ffffff', bg='#2d2d2d', bd=0)
output_text.pack()

# Function to call the recommend function with the input value and display the output
def call_recommend():
    input_value = input_entry.get()
    recommended_movies = recommend(input_value)
    recommended_movies_text = "\n".join(recommended_movies)
    output_text.delete('1.0', tk.END)
    output_text.insert(tk.END, recommended_movies_text)

# Create a button to call the recommend function with the input value and display the output
submit_button = tk.Button(root, text="Submit", font=("Arial", 14), fg='#ffffff', bg='#1db954', command=call_recommend)
submit_button.pack(pady=10)

# Center the window on the screen
root.eval('tk::PlaceWindow %s center' % root.winfo_toplevel())

# Start the main loop
root.mainloop()
```
